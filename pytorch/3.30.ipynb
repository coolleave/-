```python
# 矩阵运算 相乘
narray1 = np.random.randint(0,10,size = (3,3))
narray2 = np.random.randint(0,10,size = (3,3))
display(narray1,narray2)
```

```python
np.dot(narray1,narray2)
```

```python
narray3 = np.linalg.inv(narray1)  # 矩阵求逆
e= np.dot(narray1,narray3)
np.dot(e,narray2)
```

```python
np.dot(narray2,e)
```
