```python
import pandas as pd
```

```python
std = pd.read_excel("student.xlsx",skiprows=4)  # 加载文件，并跳过四行
# 删除空列
std.dropna(axis="columns",how='all', inplace=True)
display(std)
```

```python
# 删除空行
std.dropna(axis=0,how='all', inplace=True)
display(std)
```

```python
std.isnull()  # 检测空格
std["分数"].notnull()  # 检测非空值
```

```python
std.loc[:,"分数"] = std["分数"].fillna(0)
std
# 等同于
std.fillna({'分数': 0})
std
```

```python
# 将姓名空缺进行填充

std.loc[:,'姓名'].ffill()
std
```

```python
std.to_excel('student_clean.xlsx',index=False)
```
