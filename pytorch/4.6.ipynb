```python
import numpy as np
import pandas as pd
```

```python
df_ratings = pd.read_csv('data/movies/ratings.csv',
                 names = "userId,movieId,rating,timestamp".split(','),
                 skiprows=1
                )
df_ratings.head()
```

```python
df_excel = pd.read_excel('data/movies/users.xlsx'
                )
df_excel.to_csv('data/movies/users.csv', index=False) 

```

```python
df_uesrs = pd.read_csv('data/movies/users.csv',
                 names = "userId,age,gender,occupation,zip-code".split(','),
                 skiprows=1
                )
df_uesrs.head()
```

```python
df_movies = pd.read_csv('./data/movies/movies.csv',
                        names = "movieId,title,genres".split(','),
                        skiprows= 1
                       )
df_movies.head()
```

```python
# merge函数 根据所指定的列key，将两个表合并成一个表

df_movies_ratings = pd.merge(df_movies,df_ratings,
                             left_on=  'movieId',right_on = 'movieId',
                             how = 'inner',
                            )
df_movies_ratings.head()
                             
```
